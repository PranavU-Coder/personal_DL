imports

```python
import tensorflow as tf
import numpy as np
```

data points

```python
xs = np.array([-1.0,0.0,1.0,2.0,3.0,4.0],dtype=float)
ys = np.array([-3.0,-1.0,1.0,3.0,5.0,7.0],dtype=float)
```

simple unilayer with Keras for the neural network

```python
model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    tf.keras.layers.Dense(units=1)
])
```

to compile the neural network , using loss and an optimizer function

```python
model.compile(optimizer='sgd',loss='mean_squared_error')
```

adding 500 training cycles for the model to learn and predict based on the datapoints we fed it and the compiler functions to take a guess

```python
model.fit(xs,ys,epochs=500)
```

prediction time

```python
prediction = model.predict(np.array([10.0]))
value = prediction.item()
print(f"{value:.5f}")
```
